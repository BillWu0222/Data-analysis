```python
import pandas as pd
from scipy import stats

# 加載資料集
player = pd.read_csv("player_cleaned.csv")

# 使用Pandas的get_dummies函數執行熱編碼，將星座轉換為虛擬變數
dummies = pd.get_dummies(player['star_sign'], prefix='star_sign')

# 將虛擬變數合併回原始資料集
player = pd.concat([player, dummies], axis=1)
plt.rcParams['font.sans-serif']=['Arial Unicode MS'] 

# 編碼後的星座欄位
encoded_star_signs = player[['star_sign_Aquarius', 'star_sign_Aries', 'star_sign_Cancer', 'star_sign_Capricorn', 'star_sign_Gemini', 'star_sign_Leo', 'star_sign_Libra', 'star_sign_Pisces', 'star_sign_Sagittarius', 'star_sign_Scopio', 'star_sign_Taurus', 'star_sign_Virgo']]

# 初始化一個空的相關性係數列表
correlations = []

# 逐個計算每個星座和身高的相關性
for column in encoded_star_signs.columns:
    correlation, p_value = stats.pointbiserialr(encoded_star_signs[column], player['height'])
    correlations.append(correlation)

# 輸出相關性係數
print("各星座和身高之間的Point-Biserial相關性係數：")
for i, column in enumerate(encoded_star_signs.columns):
    print(f"{column}: {correlations[i]}")

# 創建一個DataFrame以便於繪圖
correlation_df = pd.DataFrame({'星座': encoded_star_signs.columns, '相關性係數': correlations})

# 使用Seaborn繪製條形圖
plt.figure(figsize=(12, 6))
sns.barplot(x='星座', y='相關性係數', data=correlation_df, palette='viridis')
plt.xlabel('星座')
plt.ylabel('Point-Biserial相關性係數')
plt.title('各星座和身高之間的相關性')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
match = pd.read_csv("./match_cleaned.csv")

# 計算星座和進球數的交叉表
cross_table = pd.crosstab(index=match['away_team_goal'], columns=player['star_sign_Aquarius'])

# 執行 Chi-squared test
chi2, p_value, dof, expected = chi2_contingency(cross_table)

# 輸出結果
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p_value}")

alpha = 0.05  # 顯著性水平

if p_value < alpha:
    print("根據P-value，我們可以拒絕虛無假設，認為兩者之間存在統計上的相關性。")
else:
    print("根據P-value，我們無法拒絕虛無假設，認為兩者之間沒有統計上的相關性。")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 加載資料集


# 將身高（height）分成身高區間
player['height_category'] = pd.cut(player['height'], bins=[150, 160, 170, 180, 190, 200,220])

# 執行Chi-squared test
contingency_table = pd.crosstab(player['star_sign'], player['height_category'])
chi2, p, _, _ = chi2_contingency(contingency_table)

# 輸出結果
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p}")

alpha = 0.05  # 顯著性水平

if p < alpha:
    print("根據P-value，我們可以拒絕虛無假設，認為星座（star_sign）和身高（height_category）之間存在統計上的相關性。")
else:
    print("根據P-value，我們無法拒絕虛無假設，認為星座（star_sign）和身高（height_category）之間沒有統計上的相關性。")
```

```python
# 繪製散佈圖
plt.figure(figsize=(12, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.xlabel("身高區間")
plt.ylabel("星座")
plt.title("星座和身高之間的相關性")
plt.show()
```

```python
# 繪製分佈圖
plt.figure(figsize=(12, 6))
sns.countplot(data=player, x='height_category', hue='star_sign', palette='Set2')
plt.xlabel("身高區間")
plt.ylabel("計數")
plt.title("星座和身高區間的分佈")
plt.xticks(rotation=45)
plt.legend(title='星座', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
